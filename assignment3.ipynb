* 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.


    1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
>>> addall([1])
1

>>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
45

```python
#for ~ in 리터럴
def addall(list):
    sum = 0
    for i in range(len(list)):
        sum += list[i]
    return sum
        
print addall([1])
print addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
#리스트 내포

def addall(list):
    
    
    L = sum([list[i] for i in range(len(list))])
   
    return L
        
print addall([1])
print addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

    2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
#for ~ in 리터럴
def addallodd(list):
    sum = 0
    for i in range(len(list)):
        if (list[i]%2 == 1):
            sum += list[i]
    return sum
       
        
print addallodd([1])
print addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
#리스트 내포

def addall(list):
    
    
    L = sum([list[i] for i in range(len(list)) if i%2 == 0])
   
    return L

print addallodd([1])
print addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
primeList=[2]
isContinue=True
potenNum=2

while isContinue==True:
    potenNum+=1

    for i in primeList:

        if potenNum % i == 0:
            break
        elif i==primeList[-1]:
            primeList.append(potenNum)

    if len(primeList)==10001:
        isContinue=False
    else:
        continue

print (primeList[10000])
```

[프로젝트 오일러 문제 9] a + b + c = 1000 인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
# 0<a<b<c

def tri(n):
    for a in range(1, n):
        for b in range(2, n):
            c = n-a-b
            if((c**2) == (a**2) + (b**2)):
                return a*b*c
        
print tri(1000)
            
```
